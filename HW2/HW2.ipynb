### `Import Data`  
We import the data into multiple dataframes by their subset for later use, but also can concatenate them for current use.

```python
# Read all the sheets into a dictionary of DataFrames
excel_file = 'weld_defect_dataset.xlsx'
subsets = []

for i in range(1, 6):
    sheet_name = 'subset' + str(i)
    subset = pd.read_excel(excel_file, sheet_name=sheet_name)
    subsets.append(subset)

subsetsall = pd.concat(subsets, ignore_index=True) # Concatenate all the DataFrames into a single DataFrame
subsetsall
```
